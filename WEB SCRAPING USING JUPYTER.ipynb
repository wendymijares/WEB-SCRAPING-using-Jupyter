## Web Scraping Forbes list of the world's highest-paid athletes 2023

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Forbes_list_of_the_world%27s_highest-paid_athletes'
```

```python
requests.get(url)
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
table = soup.find_all('table')[1]
table
```

```python
table2 = soup.find('table', class_ = 'wikitable')
table2
```

```python
column_headers=table2.find_all('th')
column_headers
```

```python
th_table = [title.text.strip() for title in column_headers]
th_table
```

```python
columns_table = [title.text.strip() for title in column_headers]

print(columns_table)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = columns_table)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
